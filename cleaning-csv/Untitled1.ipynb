```python
import pandas as pd
import numpy as np
df = pd.read_csv('tb.csv')
df.head()
```

```python
df = df.melt(['country','year'],['m04','m514','m014','m1524','m2534','m3544','m4554','m5564', 'm65', 'mu', 'f04','f514','f014','f1524','f2534','f3544','f4554','f5564', 'f65', 'fu'],'sexage','cases')
```

```python
df.head()
```

```python
df['gender'] = df.sexage.str.slice(0,1) 
df['age'] = df.sexage.str.slice(1)
df.head()
```

```python
df['age'] = df['age'].map({'04' : '0-4', '514' : '5-14', '1524' : '15-24', '2534' : '25-34', '3544' : '35-44', '4554' : '45-54', '5564' : '55-64', '65' : '65+', 'u' : np.nan} )
```

```python
df['age']
```

```python
final = df.dropna(subset=['cases'])
```

```python
final.head()
```

```python
final.sort_values(['country', 'year', 'age', 'gender'])
final.head()
```

```python
final = final[['country', 'year', 'age', 'gender']]
```

```python
final.head(10)
```

```python
final.to_csv('final.csv', index=False)
```
